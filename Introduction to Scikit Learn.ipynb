```python
import numpy as np
```

```python
# Get the data ready
import pandas as pd
heart_disease = pd.read_csv("data/heart-disease.csv")
heart_disease
```

```python
# Create x (features matrix)
X = heart_disease.drop("target", axis=1)

# Create y (labels)
y = heart_disease["target"]
```

```python
# Choose the right model and hyper param meters
from sklearn.ensemble import RandomForestClassifier
clf =  RandomForestClassifier(n_estimators=100)

# we keep default hyperparameters

clf.get_params()
```

```python
# 3. Fit the model to the data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
# 80% training
# 20% testing
```

```python
clf.fit(X_train, y_train);
```

```python
# make a prediction
y_label = clf.predict(np.array([0,2,3,4]))
```

```python
X_train
```

```python
X_test
```

```python
y_preds = clf.predict(X_test)
y_preds
```

```python
y_test
```

```python
# 4. Evaluate the model on training data and test data
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

```python
# Improve a model
# try different amount of n_estimators

np.random.seed(42)
for i in range(10,100,10):
    print(f"Trying model with {i} estimators ...")
    clf = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f"Model accuracy on test set: {clf.score(X_test, y_test) * 100:.2f}%")
    # this is shortcut in python format string syntax - its round(some_number, 2)
    print("")
```

```python
# save a model and load it
import pickle

pickle.dump(clf, open("random_forest_model_1.pkl","wb"))
```

```python
# import a model from .PKL file
loaded_model = pickle.load(open("random_forest_model_1.pkl", "rb"))
loaded_model.score(X_test, y_test) # last model we ran it shows that output
```
